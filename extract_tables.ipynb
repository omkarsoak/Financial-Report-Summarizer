#### Item 7,8 Table Extraction

```python
def extract_text_between_strings(html_file, start_text, end_text, output_file):
    try:
        with open(html_file, 'r') as file:
            html_content = file.read().lower()
    except FileNotFoundError:
        print("HTML file not found.")
        return
    
    start_index = html_content.find(start_text.lower())
    if start_index == -1:
        print("Start text not found.")
        return
    
    end_index = html_content.find(end_text.lower(), start_index)
    if end_index == -1:
        print("End text not found.")
        return
    
    extracted_text = html_content[start_index + len(start_text):end_index]
    
    try:
        with open(output_file, 'w') as file:
            file.write(extracted_text)
    except:
        print("Error writing to output file.")



# extract_text_between_strings("primary-document.html", "item 7. management", "item 9", "item7_8.html")
# extract_text_between_strings("item7_8.html", "summary results", "item 8. financial", "item7_tables.html")
# extract_text_between_strings("item7_8.html", "item 8. financial", "auditor since", "item8_tables.html")
# extract_text_between_strings("item8_tables.html", "basic earnings per share", "note 3", "item8_epstable.html")



```
