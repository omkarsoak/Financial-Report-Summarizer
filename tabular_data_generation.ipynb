```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up a grid of subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(16, 16))

colors = ['#B0C5A4', '#D37676', '#EBC49F', '#F1EF99', '#A78295']

# Balance Sheet
ax = axes[0, 0]
balance_sheet = statement.T[['Cash and Cash Equivalents, at Carrying Value', 'Accounts Receivable, after Allowance for Credit Loss, Current', 'Accrued Liabilities, Current']]
balance_sheet.plot(kind='line', ax=ax, color = colors)
ax.set_title('Balance Sheet - Key Metrics', fontsize=14)
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Amount', fontsize=12)
ax.tick_params(axis='x', rotation=45)
ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', fontsize=10)

# Income Statement
ax = axes[0, 1]
income_statement = statement1.T[['Revenue from Contract with Customer, Excluding Assessed Tax',
                                 'Operating Income (Loss)',
                                 'Net Income (Loss) Attributable to Parent']]
income_statement.plot(kind='bar', ax=ax, color = colors)
ax.set_title('Income Statement - Key Metrics', fontsize=14)
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Amount', fontsize=12)
ax.tick_params(axis='x', rotation=45)
ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', fontsize=10)

# Cash Flow Statement
ax = axes[1, 0]
cash_flow_statement = df2.T[['Net Cash Provided by (Used in) Operating Activities',
                             'Net Cash Provided by (Used in) Investing Activities',
                             'Net Cash Provided by (Used in) Financing Activities']]
cash_flow_statement.plot(kind='bar', ax=ax, color = colors)
ax.set_title('Cash Flow Statement - Key Metrics', fontsize=14)
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Amount', fontsize=12)
ax.tick_params(axis='x', rotation=45)
ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', fontsize=10)

# Revenue and Expense Trends
ax = axes[1, 1]
revenue_expense = statement1.T[['Revenue from Contract with Customer, Excluding Assessed Tax',
                                'Cost of Goods and Services Sold',
                                'Operating Income (Loss)']]
revenue_expense.plot(kind='line', ax=ax, color = colors)
ax.set_title('Revenue and Expense Trends', fontsize=14)
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Amount', fontsize=12)
ax.tick_params(axis='x', rotation=45)
ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', fontsize=10)

# Asset and Liability Breakdown
ax = axes[2, 0]
asset_liability = statement.T[['Assets, Current',
                               'Other Assets, Noncurrent',
                               'Liabilities, Current',
                               'Other Liabilities, Noncurrent', 'amzn_LeaseLiabilityNoncurrent']]
asset_liability.plot(kind='bar', ax=ax, color = colors)
ax.set_title('Asset and Liability Breakdown', fontsize=14)
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Amount', fontsize=12)
ax.tick_params(axis='x', rotation=45)
ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', fontsize=10)

# Adjust spacing between subplots
fig.subplots_adjust(hspace=0.5, wspace=0.5)

plt.show()
```
